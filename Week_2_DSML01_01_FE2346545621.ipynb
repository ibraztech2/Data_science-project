2. Data Importation

```python
bf_data_df = pd.read_csv("/content/Business Funding Data.csv",encoding="latin-1")
bf_data_df.head()
```

```python
bf_data_df.info()
bf_data_df.describe()
```

```python

null_count  = bf_data_df.isnull().sum()


for col, null in null_count.items():
  if not(null > 0.25 * len(bf_data_df)):
    null_count.drop(col,inplace=True)
  else:
    pass
#null_count
```

```python
bf_data_df.drop(null_count.index,axis=1,inplace=True)
bf_data_df.drop("Source Urls",axis = 1,inplace=True)
#bf_data_df
```

```python
from sklearn.preprocessing import LabelEncoder,MinMaxScaler
columns = bf_data_df.columns
label_encoder = LabelEncoder()
normalizer = MinMaxScaler()
for column in columns:
  if not bf_data_df[column].dtype == "object":
    bf_data_df[column] = normalizer.fit_transform(bf_data_df[column].values.reshape(-1,1))
  else:
    bf_data_df[column] = label_encoder.fit_transform(bf_data_df[column])



bf_data_df
```

```python
bf_data_df["Categories"].replace(8,np.nan,inplace=True)
bf_data_df.dropna(inplace=True)
col_to_drop = ["Website Domain",	"Found At",	"Amount"]
bf_data_df.drop(col_to_drop,axis=1,inplace=True)
bf_data_df
```
